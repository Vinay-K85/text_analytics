# Text Mining

The process of finding relevant information from unstructured and semi-structured data is called `text mining`. The final output is a summarized information with useful and relevant words.

Text mining involves following steps:
- Understanding the text data
- Text Preprocessing
- Shallow parsing
- Stop words
- Stemming & Lemmatizing
- Word Cloud

#### Understanding the text data

```python
# Import the necessary packages to determine the number of words in the document
from string import punctuation
from collections import Counter

import warnings
warnings.filterwarnings("ignore")
```

```python
# Display the punctuation marks from the string module
print("Different punctuation marks in the string module are: \n", punctuation)
```

```python
# Load and read the document
text_file = open("textminingwiki.txt", "r")
text_doc = text_file.read()
```

```python
# Print the document
print(text_doc)
```

```python
# Use translate() function to remove punctuation marks from the document
# Use of lower() function to convert the text in lower case
# Use of split() function to split the document in words

word_doc = text_doc.translate(punctuation).lower().split()
print(word_doc)
print("\n")
print("The number of words in the document are: \n", len(word_doc))
```

```python
# Use Counter() function to display the most common words
word_counts = Counter(word_doc)
print("The most common words in the document are: \n")

for word in word_counts.most_common(11):
    print(word)
```

#### Text Preprocessing

```python
# Import the necessary packages to be used for text preprocessing
import nltk
import re
```

```python
# Remove unwanted characters from the document using regular expressions
text_doc_2 = re.sub(r"[a-zA-Z0-9\s]", " ", text_doc,re.I|re.A)
```

```python
# Determine the number of words after removing unwanted characters
print("Number of words after removing unwanted characters from the document: \n", len(text_doc_2))
```

```python
# Convertin all the words to lowercase for exact determination
text_doc_2 = text_doc_2.lower()
```

```python
# Removing leading and trailing spaces from the document
text_doc_2 = text_doc_2.strip()
```

```python
# Display the document
print("The document after text preprocessing: \n", text_doc_2)
```

#### Shallow parsing

```python
# Tokenize the document into tokens/ words using word_tokenize() function
nltk.download("punkt")

doc_tokens = nltk.word_tokenize(text_doc_2)
print("The tokens are: \n", doc_tokens)
```

```python
# Display the nature of words
nltk.download("averaged_perceptron_tagger")
nltk_pos_tagged = nltk.pos_tag(doc_tokens)
```

```python
# Display tokens (words) and the POS tag
print("The words are: \n", nltk_pos_tagged)
```

#### Stop words

```python
# Download stop words from nltk library
nltk.download("stopwords")
```

```python
# Storing all the stop words of English from nltk corpus
eng_stopwords = nltk.corpus.stopwords.words("english")
```

```python
# Determine the number of English stop words in corpus
print("Number of English stopwords in corpus are: \n", len(eng_stopwords))
```

```python
# Display the first 30 stopwords
print("List of first 30 stopwords are: \n", eng_stopwords[:30])
```

```python
# Filtering the tokens which are not in the list of stopwords
final_tokens = [token for token in doc_tokens
               if token not in eng_stopwords]
```

```python
final_tokens
```

```python
# Joining the filtered token using join() function
new_doc = " ".join(final_tokens)
```

```python
# Determine the number of words after removal of stopwords from the document
print("Number of words after removal of stopwords are: ", len(new_doc))
print("\n")
print("First 300 letters from the new document are: \n", new_doc[0:300])
```

#### Stemming & Lemmatizing

```python
# Importing necessary packages
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
# Download the file
nltk.download("wordnet")

ps = PorterStemmer()
lem = WordNetLemmatizer()
```

```python
# Creating a list of words having some base word
my_words = ["program", "programs", "programming", "programer", "programers"]

for word in my_words:
    print("Stemming words: \n", word, ":", ps.stem(word))
    print("Lemmatizing word: \n", word, ":", lem.lemmatize(word))
```

```python
# Use of Stemming and Lemmatizing on data
lemlist = []
pslist = []
```

```python
# Creating a list of words having same base word
for word in final_tokens:
    lemlist.append(lem.lemmatize(word, "v"))
    pslist.append(ps.stem(word))
```

```python
# Display the result after lemmatization and stemming
print("After Lemmatization: \n", lemlist[0:20])
print("\n")
print("After Stemming: \n", pslist[0:20])
```

#### Word Cloud

```python
# Importing necessary packages
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Creating a word cloud of words without lemmatization
start_all_words = " "
stopwords = set(STOPWORDS)

for word in doc_tokens:
    start_all_words = start_all_words + word + " "
    
start_wordcloud = WordCloud(width = 1000, height = 1000, background_color = "white", stopwords = stopwords,
                           min_font_size = 10).generate(start_all_words)
```

```python
# Creating a word cloud of words with lemmatization
clean_all_words = " "

for new_word in lemlist:
    clean_all_words = clean_all_words + new_word + " "
    
clean_wordcloud = WordCloud(width = 1000, height = 1000, background_color = "black", stopwords = stopwords,
                           min_font_size = 10).generate(clean_all_words)
```

```python
# Plot both the word clouds
plt.figure(1, figsize = (20,10))
plt.subplot(121)
plt.imshow(start_wordcloud)
plt.subplot(122)
plt.imshow(clean_wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
